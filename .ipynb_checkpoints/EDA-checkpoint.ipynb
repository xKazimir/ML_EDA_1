# combine filters

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & 
    (df['Year of event'] == 2020) &
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & 
    (df['Year of event'] == 2020) &
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

# remove (USA) in Event name

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

# clean up athlete age

```python
df2['athlete age'] = 2020 - df2['Athlete year of birth']
```

```python
df2.head()
```

# remove h from athlete performance

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(" ").str.get(0)
```

# drop columns : Athlete club, Athlete country, Athlete year of birth, Athlete age category

```python
df2 = df2.drop (['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

# clean up null cells

```python
df2.isna().sum()
```

```python
df2[df2['athlete age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

# check for dupes 

```python
df2[df2.duplicated() == True]
```

# reset index

reset index

```python
df2.reset_index(drop = True)
```

# fix types

```python
df2.dtypes
```

```python
df2['athlete age'] = df2['athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

# rename columns

# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed         object
# Athlete ID                     int64
# athlete age                  float64

```python
df2 = df2.rename(columns = {'Year of event' : 'year',
                             'Event dates' : 'race_day',
                             'Event name' : 'race_name', 
                             'Event distance/length' : 'race_length',
                             'Event number of finishers' : 'race_number_of_finishers',
                             'Athlete performance' : 'athlete_performance',
                             'Athlete gender' : 'athlete_gender',
                             'Athlete average speed' : 'athlete_average_speed',
                             'Athlete ID' : 'athlete_id',
                             'athlete age' : 'athlete_age',
                 })
```

```python
df2.head()
```

# reorder columns

```python
df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_average_speed']]
```

```python
df3.head()
```

# questions i want to find out from the data

```python
#Difference in speed for the 50k,50mi male to female
```

```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are the best in the 50mi race (20 + races min) (show 15)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
#what age groups are the worst in the 50mi race (10 + races min) (show 20)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>9').head(20)
```

```python
# are runners faster in summer than in winter?
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x : 'Winter' if x>11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```

```python
# 50 miler only
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
